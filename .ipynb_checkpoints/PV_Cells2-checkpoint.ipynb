```python
import pylab

import skimage.morphology as morph
import skimage.feature.texture as texture
import skimage.filters as filters
import skimage.util as utils
from skimage.restoration import denoise_bilateral
from skimage.io import imread
from skimage.feature import structure_tensor, structure_tensor_eigvals
from sklearn.model_selection import StratifiedKFold,cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier,VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.decomposition import PCA
```

```python
root ='/home/marcelo/PV_Cell/' # Caminho para imagens

fnamesF = [] # Nome das imagens PV sem defeito
with open(root+"Sem_Defeito.txt") as f:
 fnamesF = [s.strip() for s in f]

fnamesT = [] # Nome das imagens PV com defeito
with open(root+"Defeito.txt") as f:
 fnamesT =  [s.strip() for s in f]

# Abre arquivos de imagens em coloca em listas separadas
imF = [imread(root+"Sem_Defeito/"+f) for f in fnamesF]
imT = [imread(root+"Defeito/"+f) for f in fnamesT]

im = {-1:imF,1:imT}

roi_1 = [(10,25),(43,265)]
roi_2 = [(60,15),(140,280)]
roi_3 = [(156,15),(238,280)]
roi_4 = [(255,25),(290,265)]
```

```python
img = imT[50]

sc = 1.
ss = 1.
dsk = 30

l = []

for roi,i,s in zip([roi_1,roi_2,roi_3,roi_4],range(1,5),4*[(121,122)]):
 x0,y0 = roi[0]
 x1,y1 = roi[1]   
 Axx, Axy, Ayy = structure_tensor(img[x0:x1,y0:y1], sigma = 2.)
 aux = structure_tensor_eigvals(Axx, Axy, Ayy)
 print(aux[0].max(),aux[0].min(),aux[1].max(),aux[1].min()) 
 pylab.figure(i,figsize=(10,20))
 pylab.subplot(s[0])
 pylab.imshow(img[x0:x1,y0:y1],'gray')
 pylab.subplot(s[1])
 pylab.imshow(aux[0])
  #img_b = filters.gaussian(img[x0:x1,y0:y1],sigma = ss)
# img_b = denoise_bilateral(img[x0:x1,y0:y1],sigma_color = sc,sigma_spatial = ss) 
# img_c = morph.black_tophat(img[x0:x1,y0:y1],morph.disk(dsk))
# l.append(img_c)    
# pylab.figure(i,figsize=(10,20))
# pylab.subplot(s[0])
# pylab.imshow(img[x0:x1,y0:y1],'gray')
# pylab.subplot(s[1])
# pylab.imshow(img_b,'gray')
# pylab.subplot(s[2])
# pylab.imshow(img_c,'gray')
 
```

```python

def glsm_features(img):
  g = texture.greycomatrix(utils.img_as_ubyte(img), [1,2,3], [0,pylab.pi/4,pylab.pi/2,3*pylab.pi/4], 
                           levels=256, normed=True, symmetric=True)
  l = []
  for f in ['contrast','dissimilarity','homogeneity','ASM','energy','correlation']:
    l.append(texture.greycoprops(g,f).mean())
  return l

def glsm_features2(img):
  offsets = [1,3,5]
  angles =  pylab.linspace(0,pylab.pi,4)
  g = texture.greycomatrix(utils.img_as_ubyte(img),offsets,angles, 
                           levels=256, normed=False, symmetric=True)
  d = []
  for f in ['contrast','dissimilarity','homogeneity','energy','correlation']:
   d.append(pylab.array([texture.greycoprops(g,f).flatten()]).transpose())
   
  return pylab.hstack(d)

def tensor_features(img,sigma=0.25):
 Axx, Axy, Ayy = structure_tensor(img, sigma)
 aux = structure_tensor_eigvals(Axx, Axy, Ayy)
 cw = pylab.zeros(img.shape) 
 idx = pylab.where((aux[0]+aux[1]) != 0)
 cw[idx] = (aux[0][idx]-aux[1][idx])**2/(aux[0][idx]+aux[1][idx])**2
 return cw.flatten()
```

```python
s = 5.0
ss = 5.0

def FeaturesPerROI(img,r):
 x0,y0 = r[0]
 x1,y1 = r[1]
 aux = glsm_features2(img[x0:x1,y0:y1]).transpose()
 #aux = pylab.vstack([tensor_features(img[x0:x1,y0:y1],sigma=s) for s in pylab.linspace(0.1,50.0,20)])
 aux = PCA(n_components = 5,whiten=False).fit_transform(aux)
 l = aux.flatten().tolist()
 return l

X = [[],[],[],[]]
lbl = []
i = 0
for label,img_list in im.items():
    for img in img_list:
     lbl.append(label)
     print(i,' ',end='')
     for ix,roi in enumerate([roi_1,roi_2,roi_3,roi_4]):
      x0,y0 = roi[0]
      x1,y1 = roi[1]
      img_b = denoise_bilateral(img[x0:x1,y0:y1],sigma_color = sc,sigma_spatial = ss) 
      #img_c = morph.black_tophat(img_b,morph.disk(dsk))    
      X[ix].append(FeaturesPerROI(img,roi))
     i = i+1  
```

```python
x,y = pylab.hstack(X),lbl
print(x.shape)
```

```python

clf1 = RandomForestClassifier(n_estimators=50,max_features="sqrt",max_depth=None,min_samples_split=2,n_jobs=-1)
clf2 = SVC(kernel='linear',probability=True)
clf3 = LogisticRegression(max_iter=250)
clf4 = GaussianNB()
clf5 = KNeighborsClassifier(n_neighbors = 3)
clf6 = VotingClassifier(estimators=[('svc',clf2),('lrc',clf3),('gnb',clf4),('knn',clf5)],voting='soft')
```

```python
from sklearn.feature_selection import RFE
for n in range(3,x.shape[1],1):
 print(n,end='')
 rfe = RFE(estimator=clf2,n_features_to_select=n,step = 2)
 for clf in [clf2,clf3,clf4,clf5,clf6]:
  p = make_pipeline(rfe,clf)
  score = cross_val_score(clf, x,y, cv=StratifiedKFold(n_splits=10,shuffle=True),scoring='f1',n_jobs=-1)
  print(" {0:.2f}".format(score.mean())," {0:.2f}".format(score.std()),end = '')
 print()
```

```python
for k in range(3,X.shape[1],10):
 print(k,end='')
 for clf in [clf1,clf2,clf3,clf4,clf5,clf6]:  
  pl = make_pipeline(PCA(n_components=k,whiten=False),clf)
  score = cross_val_score(pl,X,y, cv=StratifiedKFold(n_splits=10,shuffle=True),scoring='f1',n_jobs = -1)
  print(" {0:.2f}".format(score.mean())," {0:.2f} ".format(score.std()),end = '')
 print()   
```
